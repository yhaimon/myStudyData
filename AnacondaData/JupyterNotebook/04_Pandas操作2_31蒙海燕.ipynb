列添加通过对df[新的列标签]赋值实现。
列选择可以使用del df[列标签]或者df.pop(列标签)实现。


```python
#导入pandas库，并取一个别名pd
import pandas as pd
d1 = {'name':['小明','小花','小兰','小胜'],  
     'gender':['男','女','女','男'],
     'age':[20,22,19,23],
     'clazz': ['1班','1班','2班','1班']}
df1 = pd.DataFrame(d1)
print('原数据帧：')
print(df1)
```

```python
#列选择：df[列标签] 或者df[列标签列表]（注意是二维列表！！！）
print('选择name列：')
print(df1["name"])
print('选择name,age列：')
print(df1[["name","age"]])
```

```python
#列添加：df[新的列标签] = pd.Series(新数据列表,index=新数据对应的索引列表)
#添加city列，数据有北京、上海、深圳，索引分别为0、1、3
print("添加city列：")
df1['city']=pd.Series(['北京','上海',"深圳"],index=[0,1,3])
print(df1)
#列删除：del df[要删除的列标签] 或者 数据结构名.pop(要删除的列标签)
#将刚才添加的city列删除
print("删除city列：")
#del df1["city"]
df1.pop("city")
print(df1)
```

行操作包括行选择、行添加和行删除。  
行的选择可以按行标签、整数位置进行选择，函数分别是loc和iloc函数，另外还可以使用行索引或切片进行选择。  
函数loc的参数是行标签，iloc的参数是行号。  
行添加通过df.append函数实现；行删除通过df. drop函数实现。  

```python
d2 = {'name':['小明','小花','小兰','小胜'],  
   'gender':['男','女','女','男'],
     'age':[20,22,19,23],
    'clazz': ['1班','1班','2班','1班']}
df2 = pd.DataFrame(d2,index=['first','second','third','fourth'])
print('原数据帧：')
print(df2)
#单行选择：df.loc[行标签名字] 
#例如：选择second行
print('\n选择second行：')
print(df2.loc["second"])

#多行选择：df.loc[行标签列表] 注意是二维列表！！！ 
#例如：选择second、first行
print('\n选择second、first行：')
print(df2.loc[["first","second"]])

#单行选择：df.iloc[行下标] 注意第一行下标是从0开始的！！
#例如选择第一行
print('\n选择第一行：')
print(df2.iloc[0])
#多行选择：df.iloc[[选中的第1行下标,选中的第2行下标,....]] 注意是二维列表！！！
print('\n选择0，2行：')
print(df2.iloc[[0,2]])
#多行选择：df.iloc[start:end] 选中[start:end)下标范围上的行
print('\n选择0:2行：')
print(df2.iloc[0:2])
```

行添加：df[*新的行标签*] = pd.Series(*新数据列表*,*index=新数据对应的索引列表*)

```python
#例如：添加新的一行，数据有小红、女、27、6班 索引分别对应name、gender、age、clazz
#新的数据结构名 = 老的数据结构._append(新的一行,ignore_index=True)       *****python2中使用.append ，python3中使用._append *******
#这里建议先忽略掉会报的future warning的错
print("\n原来：df1:\n",df1)
print('添加s行后的df1：')
s = pd.Series({'name':'小红', 'gender':'女', 'age':27, 'clazz':'6班'})
#print(s)
df1 = df1._append(s,ignore_index=True)
print(df1)
#行删除：新的pandas数据结构名字 = 旧结构名.drop(待删除的行对应的索引)
print("删除第二行后,df1：\n")
df1 = df1.drop(1)
print(df1)
```

Pandas有两种排序方式，它们分别是：按标签排序（*sort_index*函数）和按实际值（*sort_values*函数）排序。


```python
#排序操作
import numpy as np
data = np.array([[2,5,3,7],[16,14,2,30],[29,27,2,25]])
print(data)
df6 = pd.DataFrame(data,index=[0,4,2],columns=['a','c','b','d'])
print('\n原数据帧：')
print(df6)
print("--"*8)
#行索引升序 数据结构名.sort_index()
print('按行索引升序：')
print(df6.sort_index())
print("--"*8)
#行索引降序 数据结构名.sort_index(ascending=False)
print('按行索引降序：')
print(df6.sort_index(ascending=False))
print("--"*8)
#列索引升序 数据结构名.sort_index(axis=1)
print('按列索引升序：')
print(df6.sort_index(axis=1))
print("--"*8)
#列索引降序 数据结构名.sort_index(ascending=False,axis=1)
print('按列索引降序：')
print(df6.sort_index(ascending=False,axis=1))
print("--"*8)
#按d列排序 数据结构名.sort_values(by=列名)
print('按d列排序：')
print(df6.sort_values(by='d'))
print("--"*8)
```

```python
#遍历操作
data2 = np.array([[2,5,3,7],[16,14,2,16],[29,27,2,25]])
df7 = pd.DataFrame(data2,index=[0,4,2],columns=['a','c','b','d'])
print(df7)
'''
for i,j in dict.items():
    print(i,j)
在pandas数据结构中 items()需替换为iteritems()    
'''
#类似于字典形式,遍历df7                                             ***iteritems()已被弃用，应使用items****
for x,y in df7.items():
    print(x, y.values)
```

<font color=Tomato>**第三部分**</font>-**DataFrame分组与聚合**  

```python
d6 = {'name':pd.Series(['小明','小花','小兰','小胜']),  
     'gender':pd.Series(['男','女','女','男']),
     'age':pd.Series([20,22,19,23]),
     'height':pd.Series([178,165,163,175]),
     'class': pd.Series(['1班','1班','2班','1班'])}
df8 = pd.DataFrame(d6)
print('原数据帧：')
print(df8)
```

<font color = tomato>**分组**</font> *groupby*的过程就是将原有的*DataFrame*按照*groupby*的字段，划分为若干个分组*DataFrame*，被分为多少个组就有多少个分组*DataFrame*。  
所以说，在*groupby*之后的一系列操作（如agg、apply等），均是基于子*DataFrame*的操作。

```python
#分组
#单个分组格式为：df.groupby(字段名).get_group(字段属性值)
print('男生组：')
print(df8.groupby("gender").get_group("男"))
print('--'*8)
print('女生组：')
print(df8.groupby("gender").get_group("女"))
print('--'*8)

#多个分组格式为：df.groupby(字段名列表).groups
print('按班级、性别分组：')
print(df8.groupby(["class","gender"]).groups)
print('--'*8)
```

<font color=Tomato>**聚合函数**</font>为每个组返回单个聚合值。当创建了分组*(group by)*对象，就可以对分组数据执行多个聚合操作。  

```python
d7 = {'name':pd.Series(['小明','小花','小兰','小胜']),  
   'gender':pd.Series(['男','女','女','男']),
     'age':pd.Series([20,22,19,23]),
    'height':pd.Series([178,165,163,175]),
    'class': pd.Series(['1班','1班','2班','1班'])}
df9 = pd.DataFrame(d7)
groups = df9.groupby('class')      #  df9.groupby(by=["class"])
print(df9)

print('--'*8)
print('班级平均：') #格式为: 分组名.agg(np.mean) 
print(groups[["height"]].agg(np.mean))

print('不同性别学生的年龄最大值：') #格式为: df.groupby(属性).agg(np.max)
print(df9.groupby(by=["gender"])[["age"]].agg(np.max) )

print('--'*8)
print('不同性别学生人数：') #np.size表示元素数量
print(df9.groupby(by=['gender'])[['gender']].agg(np.size) )
```
