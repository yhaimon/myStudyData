```python
from pyecharts.charts import Bar,Line,Pie,Tab
from pyecharts import options as opts
from pyecharts.globals import ThemeType
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r'5月14日\gapminder.tsv',sep='\t',header=0)
data.head()
```

```python
a=data.groupby("continent")["country"].count()
a
```

```python
bar = Bar(init_opts = opts.InitOpts
          (width='800px',
           height='600px',
           theme = ThemeType.VINTAGE,
           animation_opts=opts.AnimationOpts(
                          animation = True,
                          animation_duration =1000,
                          animation_delay = 300,
                          animation_easing = 'cubicOut'))
         )
bar.add_xaxis(a.index.tolist())
bar.add_yaxis(series_name = "国家数量",
             y_axis = a.values.tolist())
bar.set_global_opts(title_opts = opts.TitleOpts(
                    title = "不同大陆国家数量"),
                    toolbox_opts = opts.ToolboxOpts(is_show=True),
                    datazoom_opts = opts.DataZoomOpts(is_show=True,
                                                     type_='slider')
                    )
```

```python
bar.render_notebook()
```

```python
b=data.groupby("year")["gdpPercap"].mean()
b
```

```python
line = Line(init_opts = opts.InitOpts
          (width='600px',
           height='400px',
           theme = ThemeType.VINTAGE,
           animation_opts=opts.AnimationOpts(
                          animation = True,
                          animation_duration =1000,
                          animation_delay = 300,
                          animation_easing = 'cubicOut'))
         )
line.add_xaxis(b.index.tolist())
line.add_yaxis(series_name = "平均GDP",
             y_axis = np.round(b.values,2).tolist())
line.set_global_opts(title_opts = opts.TitleOpts(
                    title = "不同时期平均GDP"),
                    toolbox_opts = opts.ToolboxOpts(is_show=True),
                    datazoom_opts = opts.DataZoomOpts(is_show=True,
                                                     type_='slider')
                    )
```

```python
line.render_notebook()
```
