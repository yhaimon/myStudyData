# 导包

```python
# 导包
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

# 读取数据

```python
# 读取数据
data = pd.read_csv('C:/Users/MACHENIKE/Desktop/gapminder.tsv', sep='\t')
```

# Q1. 请用合适图表表示2007年所有大陆的频数（每个大陆国家数量）情况，哪个国家的频数最高？

```python
# Q1. 大陆国家频数情况
continent_counts = data[data['year'] == 2007]['continent'].value_counts()

plt.figure(figsize=(8, 6))
continent_counts.plot(kind='bar', color='skyblue')
plt.title('2007年各大陆国家频数情况')
plt.xlabel('大陆')
plt.ylabel('频数')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Q2. 请用合适图表表示2007年不同大陆的GDP（gdpPercap）平均情况。

```python
# Q2. 不同大陆的平均GDP情况
continent_gdp = data[data['year'] == 2007].groupby('continent')['gdpPercap'].mean()

plt.figure(figsize=(8, 6))
continent_gdp.plot(kind='bar', color='lightgreen')
plt.title('2007年不同大陆的平均GDP情况')
plt.xlabel('大陆')
plt.ylabel('平均GDP')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Q3. 请用合适图表 表示中国从1952至2007年 与 人均寿命 情况。

```python
# Q3. 中国从1952至2007年与人均寿命情况
china_data = data[data['country'] == 'China']
plt.plot(china_data['year'], china_data['lifeExp'])
plt.title('中国人均寿命 (1952-2007)')
plt.xlabel('年份')
plt.ylabel('人均寿命')
plt.show()
```

# Q4. 请用合适的图表表示中国的人口与人均GDP情况，阿富汗人口与人均GDP情况，两者有区别吗？

```python
# Q4. 中国人口与人均GDP情况，阿富汗人口与人均GDP情况
china_pop_gdp = data[data['country'] == 'China'][['year', 'pop', 'gdpPercap']]
afghanistan_pop_gdp = data[data['country'] == 'Afghanistan'][['year', 'pop', 'gdpPercap']]
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(china_pop_gdp['year'], china_pop_gdp['pop'], label='China')
plt.plot(afghanistan_pop_gdp['year'], afghanistan_pop_gdp['pop'], label='Afghanistan')
plt.title('中国与阿富汗人口变化 (1952-2007)')
plt.xlabel('年份')
plt.ylabel('人口')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(china_pop_gdp['year'], china_pop_gdp['gdpPercap'], label='China')
plt.plot(afghanistan_pop_gdp['year'], afghanistan_pop_gdp['gdpPercap'], label='Afghanistan')
plt.title('中国与阿富汗人均GDP变化 (1952-2007)')
plt.xlabel('年份')
plt.ylabel('人均GDP')
plt.legend()
plt.show()
```

# Q5. 请用折线图对比比较美国('United States')， 加拿大('Canada')与巴西('Brazil')历年来的人均GDP情况，三条折线作在同一个图中。

```python
# Q5. 美国、加拿大和巴西历年来的人均GDP情况
us_data = data[data['country'] == 'United States']
canada_data = data[data['country'] == 'Canada']
brazil_data = data[data['country'] == 'Brazil']
plt.plot(us_data['year'], us_data['gdpPercap'], label='United States')
plt.plot(canada_data['year'], canada_data['gdpPercap'], label='Canada')
plt.plot(brazil_data['year'], brazil_data['gdpPercap'], label='Brazil')
plt.title('美国、加拿大和巴西历年人均GDP情况')
plt.xlabel('年份')
plt.ylabel('人均GDP')
plt.legend()
plt.show()
```
