```python
import pandas as pd
import numpy as np
from pyecharts.charts import Bar,Line,Timeline,Page,Map,Pie,Tab
from pyecharts import options as opts

import matplotlib.pyplot as plt
import seaborn as sns
from pylab import mpl
from pyecharts import options as opts
from pyecharts.globals import ThemeType

# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

```python
df = pd.read_csv(r'2024世界GDP数据分析.csv',
                index_col = 0,encoding ='gbk')
df.head()
```

```python
df_world_t = pd.DataFrame(
                        data = df.values.T,
                        index = df.columns,
                        columns = df.index)
df_world_t.header()
```

```python
tl_bar = Timeline(init_opts= opts.InitOpts(width='900px',
                                          height='500px'))
```

```python
df_world_t.head()
```

```python
pd.DataFrame(df_world_t[1960]).sort_values(by=1960)
```

```python
for year in range(1960,2023):
    # 1. 构造每年的数据集
    df_year_bar = pd.DataFrame(df_world_t[year]).sort_values(by=year)
    # 2. 定义X,Y轴数据
    x_list = df_year_bar.index.to_list()
    y_list = df_year_bar[year].to_list()
    # 3. 绘制Bar图
    bar = (Bar()
            .add_xaxis(x_list)
            .add_yaxis(series_name="",
                      y_axis = y_list,
                      category_gap= '50%')
            .reversal_axis()
            .set_global_opts(title_opts=opts.TitleOpts('{}年各国GDP情况'.format(year)))
          )
    tl_bar.add(chart = bar, time_point='{}年'.format(year))
    tl_bar.add_schema(play_interval =300, is_auto_play = False)      
```

```python
tl_bar.render_notebook()
```
