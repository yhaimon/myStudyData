```python
from pyecharts import options as opts
from pyecharts.charts import Bar,Line,Pie,Tab
from pyecharts.globals import ThemeType
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'5月14日\tip.csv')
df
```

```python
df["day"].unique().tolist()
```

```python
df["day"]
#count/sum/mean/median/max/min
y_line = df.groupby('day')['total_bill'].sum()
print(y_line)
#重置数据帧的索引，并使用默认索引。如果数据帧具有多重索引，则此方法可以删除一个或多个level。
y_line = y_line.reset_index()
print(y_line)
```

```python
#创建一个字典
day_num = {
    "Sun":7,
    "Sat":6,
    "Fri":5,
    "Thur":4,
    "Wed":3,
    "Tue":2,
    "Mon":1
}
```

```python
#排序
df['day_num'] = df['day'].map(day_num)
df.head()
```

```python
#分组
y_line = df.groupby('day_num')['total_bill'].sum()
y_line = y_line.reset_index()
y_line
```

```python
line = Line(init_opts = 
          opts.InitOpts(width='750px',
                        height="500px",
                       theme = ThemeType.VINTAGE,
                        #动画
                        animation_opts= opts.AnimationOpts(
                                        animation = True,                                        
                                        animation_duration = 1000,#持续时间
                                        animation_delay = 300,#延迟
                                        animation_easing = "elasticOut"#动画方式
                                    )
                        )
         )

#line.add_xaxis(df["day"].unique().tolist())
line.add_xaxis(['星期一','星期二','星期三','星期四','星期五','星期六','星期日',])
line.add_yaxis(series_name = "消费总额",
             #y_axis = y_line['total_bill'].tolist()
               y_axis = np.round(y_line['total_bill'],2).tolist()#np.round保留两位小数
             )
line.set_global_opts(title_opts =
                    opts.TitleOpts(
                        title = "消费总额与日期情况"
                        ),                    
                     toolbox_opts = opts.ToolboxOpts(is_show =True),#工具箱
                     datazoom_opts = opts.DataZoomOpts(is_show=True,type_='slider')#数据缩放
                    )
```

```python
line.render("5月14日\消费.html")
line.render_notebook()
```

```python
#饼图

pie = Pie(init_opts = 
          opts.InitOpts(width='750px',
                        height="500px",
                       theme = ThemeType.VINTAGE,
                        #动画
                        animation_opts= opts.AnimationOpts(
                                        animation = True,                                        
                                        animation_duration = 1000,#持续时间
                                        animation_delay = 300,#延迟
                                        animation_easing = "elasticOut"#动画方式
                                    )
                        )
         )
#pie 数据添加需要打他 pair[(男，200),(女,200)]
pie.add("",[('male',225),("female",118)])

pie.set_global_opts(title_opts =
                    opts.TitleOpts(
                        title = "消费总额与日期情况"
                        ),                    
                     toolbox_opts = opts.ToolboxOpts(is_show =True),#工具箱
                     datazoom_opts = opts.DataZoomOpts(is_show=True,type_='slider')#数据缩放
                    )

```

```python
pie.render_notebook()
```

```python
tab = Tab()
tab.add(line,"消费纵隔与日期")
tab.add(pie,"顾客性别比")
tab.render_notebook()
```
