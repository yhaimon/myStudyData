# 四：代码实现

```python
from pyecharts.charts import Bar, Timeline
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
import pandas as pd
import numpy as np 
from pyecharts.globals import ThemeType, SymbolType  
  
```

```python
df = pd.read_csv('5月21日\\2024世界GDP数据分析.csv',
 encoding='gbk',index_col=0)
df.tail()
```

```python
#为了后面循环构造数据方便，需要重构成以下数据集
df_new= pd.DataFrame(
 index=df.columns,
 columns=df.index,
 data = df.values.T)
df_new
```

```python

#构造一个升序的数据集
df_2023 =pd.DataFrame(df_new[2023].sort_values())
df_2023
```

```python

# 构造2023条形图的xy轴，x轴为国家分类，y轴为GDP数据
x_2023_data = df_2023.index.to_list()
y_2023_data =df_2023[2023].to_list()
y_2023_data
x_2023_data
```

```python
# 绘制图形
bar_2023new = (
 Bar(init_opts = 
          opts.InitOpts(width='900px',
                        height="500px",
                       theme = ThemeType.VINTAGE,
                        #动画
                        animation_opts= opts.AnimationOpts(
                                        animation = True,                                        
                                        animation_duration = 1000,#持续时间
                                        animation_delay = 300,#延迟
                                        animation_easing = "elasticOut"#动画方式
                                    )
                        ))
 .add_xaxis(x_2023_data)
 .add_yaxis(series_name="",y_axis=y_2023_data)
 .reversal_axis()
 .set_global_opts(
      tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"), 
     title_opts=opts.TitleOpts("2023年世界经济排名"),
     toolbox_opts = opts.ToolboxOpts(is_show =True),#工具箱
     datazoom_opts = opts.DataZoomOpts(is_show=True,type_='slider'),#数据缩放 
     visualmap_opts=opts.VisualMapOpts(
            # 是否显示视觉映射配置
            is_show = True,type_ = "color",min_ = 0,
            max_ = 200,
            range_text = ['高', '低'],
            range_color = ['#32CD32','#00FF7F','#87CEFA','#48D1CC','#BA55D3','#DB7093'],
            range_size = None,
            range_opacity = None,            
    )
                 )
)
bar_2023new.render_notebook()
```

```python
bar_2023new.render("2023年世界经济排名.html")
```

```python
tl = Timeline(init_opts=opts.InitOpts(width='1400px', height='800px'))
```

```python
for i in range(1960, 2024):
    # 构造1年的数据
    df_year = pd.DataFrame(df_new[i].sort_values())
    x_year_data = df_year.index.to_list()
    y_year_data = df_year[i].to_list()

    # 创建Bar对象
    bar_year = (
        Bar()
        .add_xaxis(x_year_data)
        .add_yaxis(
            series_name="",
            y_axis=y_year_data,
            itemstyle_opts=opts.ItemStyleOpts(color=JsCode("""
                function (params) {
                    var colorList = ['#F08080', '#FFD700','#FF6347', '#FFA500', '#BDB76B',  '#C0C000', '#efa18d', '#9ACD32', '#87CEEB', '#48D1CC'];
                    return colorList[params.dataIndex % colorList.length];
                }
            """))  # 使用JsCode定义颜色函数
        )
        .reversal_axis()  # 反转坐标轴
        .set_global_opts(title_opts=opts.TitleOpts("{}年世界经济排名".format(i)))
    )

    tl.add(bar_year, "{}年".format(i))  # 将Bar对象添加到Timeline
    tl.add_schema(play_interval=300, is_auto_play=False)  # 添加播放控制选项


# In[49]:


tl.render('1960-2023年世界GDP动态排名.html')
```

     画图过程。首先，我们使用add_schema()方法为时间线设置样式，如时间轴类型、方向、播放速度等。接着，我们使用add()方法添加图表和时间点。每个时间点对应一个GDP数据的图表，图表类型可以选择柱状图、折线图等。我们为每个时间点的图表设置相应的x轴（如年份或季度）和y轴（GDP数值），并添加必要的标题、图例等元素。     最后，我们使用render()方法将绘制好的时间轮播图渲染为一个HTML文件，并可以通过浏览器打开查看即可。
