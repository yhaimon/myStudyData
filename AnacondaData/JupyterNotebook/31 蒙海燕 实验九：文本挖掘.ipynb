```python
df_path = open(r"D:\cxdownload\实验12 商品评价信息.csv")
df_data = pd.read_csv(df_path)
df_data.shape
```

```python
df_c = df_data.drop_duplicates()
df_c.head()
```

```python
str_cm = str(df_c['评价信息'].values)
str_cm
```

```python
cut_words= jieba.lcut(str_cm)
cut_words[0:10]
```

```python
file_p = open(r'D:\cxdownload\停用词表.txt',encoding='utf-8')
stop_words = file_p.read()
```

```python
new_data = []
for word in cut_words:
    if word not in stop_words:
        new_data.append(word)
        
new_data  
```

```python
freq_list = FreqDist(new_data)
freq_list
data_list = freq_list.most_common()
data_list
```

```python
from pyecharts.charts import WordCloud
from pyecharts import options as opts
```

```python
wordcloud = (WordCloud()
             .add(series_name="",
                  data_pair=data_list,
                  shape = 'star',
                  word_size_range=[20,70],)
                  .set_global_opts(title_opts=
                                  opts.TitleOpts(
                                  title="评价词云图"))
                 )
wordcloud.render_notebook()
```
